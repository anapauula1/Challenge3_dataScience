```python
df = pd.read_csv('TelecomX_Data.csv')
```

```python
#eliminacion de columnas irrelevantes
df = df.drop(columns=['customerID']) # como ejemplo esta la columna del id
```

```python
# aplico a todas las columnas categoricas
columnasAplicadas = pd.get_dummies(df, drop_first=True)
# trasformo las categoricas en numericas
```

```python
columnasAplicadas.head()
```

```python
columnasAplicadas.columns
```

```python
columnasAplicadas["Churn"] = df["Churn"].map({"Yes": 1, "No": 0})
```

```python
matrix = columnasAplicadas.corr()
```

```python
matrix
```

```python
from sklearn.model_selection import train_test_split

# X: todas las columnas menos la variable objetivo
X = columnasAplicadas.drop('Churn', axis=1)
# y: variable objetivo
y = columnasAplicadas['Churn']

# División 70% entrenamiento / 30% prueba, estratificada
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=42, stratify=y
)
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, classification_report

escala = StandardScaler()
X_train_scaled = escala.fit_transform(X_train)
X_test_scaled = escala.transform(X_test)

# creo el modelo
modelo = LogisticRegression(random_state=42)
modelo.fit(X_train_scaled, y_train)

# Predecir
predec = modelo.predict(X_test_scaled)
```
